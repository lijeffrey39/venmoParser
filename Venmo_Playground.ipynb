```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
df = pd.read_csv('venmoParser.csv')
```

```python
positives = 0
negatives = 0
jeffPos = 0
jeffNeg = 0
```

```python
for index, row in df.iterrows():
    if ('Jeffrey Li' in row['0']):
        if (row['1'] == '+'):
            positives += row['2']
        else:
            negatives -= row['2']
```

```python
# Make a fake dataset:
height = [positives, negatives]
bars = ('Cash from Jeffrey', 'Cash to Jeffrey')
y_pos = np.arange(len(bars))
 
# Create bars
barlist = plt.bar(y_pos, height)
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
plt.ylabel('Money')
plt.title('Interactions with Jeffrey Li')
 
# Show graphic
barlist[1].set_color('r')
plt.show()
```
